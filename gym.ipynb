```
import numpy as np
import matplotlib.pyplot as plt
from IPython import display as ipythondisplay
```

```
import gym
env= gym.make('Blackjack-v1')
```

```
for i_episode in range(100):
  #Khởi tạo môi trường và trạng thái ở mỗi vòng lặp
  state = env.reset()
  while True:
    print(state)
    #Chọn hành động ngẫu nhiên (Hit hoặc Stand ) 
    action = env.action_space.sample()
    #Thực hiện action đã chọn và trả về các kết quả (state, phần thưởg, done là trò chơi đã end chưa, info thông tin bổ sung về mtr nếu có)
    state, reward, done, infor, = env.step(action)
    if done:
      print("End game! Reward: ", reward)
      print("You won \n") if reward >0 else print('You lost \n')
      break
```

```
for i_episodes in range(400):
  #reset lại điểm
  state = env.reset()
  episode = []
  while True:
    #nếu điểm lớn 18 thì action = 0 là dừng = 1 là uýnh bài
    action = 0 if state[0] >18 else 1
    #Thực hiện action đã chọn và trả về các kết quả (state, phần thưởg, done là trò chơi đã end chưa, info thông tin bổ sung về mtr nếu có)
    next_state, reward, done, infor = env.step(action)
    #episode là list để lưu trữ điểm, action và reward
    episode.append((state,action,reward))
    #cập nhật lại điểm
    state = next_state
    if done:
      print("End game! Reward: ", reward)
      print("You won: \n") if reward >0 else print('You lost \n')
      break
```
